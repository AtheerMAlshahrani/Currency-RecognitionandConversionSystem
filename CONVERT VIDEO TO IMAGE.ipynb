```python
import cv2
import os
```

```python
cam_money = cv2.VideoCapture("C:\\VIDEO\\1.mp4")
```

```python
try:
    # creating a folder named data
    if not os.path.exists('data_money'):
        os.makedirs('data_money')
except OSError:
    print ('Error: Creating directory of data_money')
currentframe_money = 0
while(True):
    ret_money,frame_money = cam_money.read()
    if ret_money:
        name_money = 'data_money/money' + str(currentframe_money) + '.jpg'
        print ('Creating...' + name_money)        
        cv2.imwrite(name_money, frame_money)  
        currentframe_money += 1
    else:
        break
cam_money.release()
cv2.destroyAllWindows()
```
